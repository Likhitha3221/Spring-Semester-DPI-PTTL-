```python
import requests
import json
import pandas as pd
import re

# Your Jooble API key
API_KEY = "eb5c2773-f6e1-401f-99e4-c93264ba1582"

# Jooble API URL
url = f"https://jooble.org/api/{API_KEY}"

# Define expanded search parameters to get more results
payload = {
    "keywords": "Technology",
    "location": "Chicago",
    "page": 1,  # Start with page 1
    "searchMode": 1,  # Optional: Set to 1 for broader searches
    "radius": 50,  # Search radius in miles
    "results_per_page": 50  # Increase number of results per page (if supported)
}

# Set headers
headers = {'Content-Type': 'application/json'}

# List to store all filtered jobs
all_tech_jobs = []

# Predefined list of common tech skills to extract
skills_list = ["Python", "Java", "JavaScript", "SQL", "C#", "C++", "Ruby", "AWS", "Azure", "Docker", "Kubernetes",
               "React", "Node.js", "DevOps", "Machine Learning", "AI", "Data Science", "Big Data", "Cloud",
               "Tableau", "Power BI", "Agile", "Scrum", "Cybersecurity", "Networking", "Linux", "Docker"]

# Loop through multiple pages to get more results (e.g., 5 pages for a wider search)
for page in range(1, 6):  # 5 pages to start with
    payload['page'] = page
    response = requests.post(url, data=json.dumps(payload), headers=headers)
    job_data = response.json()

    # Filter jobs based on technology-related titles
    tech_keywords = ["Software", "Engineer", "Developer", "Tech", "AI", "Cloud", "Cybersecurity",
                     "IT", "Data", "Machine Learning", "Analytics", "Robotics", "Infrastructure", "Full Stack", "DevOps", "Data Engineering"]

    tech_jobs = [job for job in job_data.get("jobs", []) if any(word in job['title'] for word in tech_keywords)]

    # Add to the list of all tech jobs
    all_tech_jobs.extend(tech_jobs)

# Function to extract skills from the job description or title
def extract_skills(job):
    skills_found = []

    # Check if description exists and extract skills from it
    description = job.get('description', '').lower()
    for skill in skills_list:
        if skill.lower() in description:
            skills_found.append(skill)

    # If no skills found in description, check the title
    if not skills_found and 'title' in job:
        title = job['title'].lower()
        for skill in skills_list:
            if skill.lower() in title:
                skills_found.append(skill)

    return ", ".join(skills_found) if skills_found else "Not Provided"

# Convert job data to DataFrame
df = pd.DataFrame(all_tech_jobs)

# Add skills to the DataFrame by applying the extract_skills function to each job
df['skills'] = df.apply(lambda job: extract_skills(job), axis=1)

# Fill missing "industry" values with a placeholder ("Not Provided")
df['industry'] = df.get('industry', 'Not Provided')

# Select relevant columns (checking if 'industry' and 'skills' are available)
df = df[['title', 'company', 'location', 'industry', 'skills', 'link']]

# Save to CSV
csv_filename = "expanded_technology_jobs_chicago_with_skills.csv"
df.to_csv(csv_filename, index=False)

print(f"Filtered job data with industry and skills info saved as {csv_filename}")
```
